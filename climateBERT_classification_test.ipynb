```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

```python
# This loads both climateBERT from HuggingFace and a tokenizer

model_name = "climatebert/distilroberta-base-climate-f"
tokeniser = AutoTokenizer.from_pretrained(model_name, local_files_only=True)
model = AutoModelForSequenceClassification.from_pretrained(model_name, local_files_only=True)
```

```python
classifier = pipeline("text-classification", model=model_name, local_files_only=True)
```

```python
# Simple task of classifying whether a text is climate related or not

texts = ["We are committed to sustainability by reducing CO2 emissions by 2030",
         "There is a large body of evidence for climate change"
         "I enjoy looking at trees"
         "This is my house and I want to make it pretty by 2025"
]

results = classifier(texts)

for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Label: {result["label"]}, Score: {result["score"]}\n")
```
