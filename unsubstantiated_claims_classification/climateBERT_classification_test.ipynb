```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

```python
# This loads both climateBERT from HuggingFace and a tokenizer

model_name = "climatebert/distilroberta-base-climate-detector"
tokeniser = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
classifier = pipeline("text-classification", model=model_name)
```

```python
# Simple task of classifying whether a text is climate related or not

texts = ["We are committed to sustainability by reducing CO2 emissions by 2030",
         "There is a large body of evidence for climate change",
         "I enjoy looking at trees",
         "By 2025, we will have a dog and it will be a day",    # Add 'warm' before 'day' to make it climate related accoding to the model
         "Hello, this is an unrelated sentence",
         "Warmth"
]

results = classifier(texts)

for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Label: {result["label"]}, Score: {result["score"]}\n")
```
